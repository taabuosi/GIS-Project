```python
# Step 2: Inspect the first few rows to understand the data structure
cities = gpd.read_file(r"C:\Users\taabuosi\Downloads\USA (1)\CITIES.SHP")
print(cities.columns)  # Print column names to check available attributes
print(cities.head()) # Preview the first few rows
```

```python
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
```

```python
# Step 3: Plot the cities based on 'POP1990' (population in 1990)
fig, ax = plt.subplots(figsize=(12, 12), constrained_layout=True)
cities.plot(column='POP1990', ax=ax, legend=True, cmap='coolwarm', 
            legend_kwds={'label': "Population in 1990", 'orientation': 'horizontal'})
# Step 4: Customize the plot layout
ax.set_title("Cities in the USA (Colored by Population in 1990)", fontsize=16)
ax.set_xlabel("Longitude")
ax.set_ylabel("Latitude")
x, y = 0.9, 0.9  # position in the plot (in normalized coordinates)
ax.annotate('N', xy=(x, y), xycoords='axes fraction', fontsize=20,
            horizontalalignment='center', verticalalignment='center', color='black', fontweight='bold')
scale_bar = inset_axes(ax, width="5%", height="5%", loc='lower right')
scale_bar.set_facecolor('white')
scale_bar.set_xticks([])
scale_bar.set_yticks([])
scale_bar.set_xlim(0, 100)
scale_bar.set_ylim(0, 10)
scale_bar.plot([0, 50], [5, 5], color="black", lw=4)  # scale bar (adjust length for your data)
scale_bar.text(55, 5, "50 m", verticalalignment='center', horizontalalignment='left', fontsize=12)

# Step 5: Show the plot
plt.show()
```
